```python
#from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt 
import matplotlib 
from pylab import *
import numpy as np 
import random

%matplotlib inline
```

```python
# Número de pares atributos (x) e labels (y)
M = 1000
```

```python
# Valores de entradas exemplos, features ou atrubutos

x1 = 2 * np.random.rand(M, 1)
y = 2 + 2 * x1 #+ np.random.randn(M, 1)
```

```python
# Gerando valores para os parâmetros
N = 200 
a0 = np.linspace(-10.0, 14.0, N)
a1 = np.linspace(-10.0, 14.0, N)
A0, A1 = np.meshgrid(a0, a1) 
```

```python
# Gerando os ponto para plotar a superfície de erro 
J = np.zeros((N,N))
for iter1 in range(0,N):
    for iter2 in range(0,N):
        yhat = A0[iter1][iter2] + A1[iter1][iter2]*x1
        J[iter1][iter2] = (1/M)*np.sum(np.square(y-yhat))
```

```python
# Plotandoa a superfície de erro 
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')

surf = ax.plot_surface(A0, A1, J, cmap=cm.coolwarm, linewidth=0, antialiased = False)

fig.colorbar(surf, shrink=0.5, aspect =5)

ax.set_xlabel('$a_0$')
ax.set_ylabel('$a_1$')
ax.set_zlabel('$J_e$')

plt.title('Função de custo da superfície')
ax.view_init(20,45)

```

```python
X = np.c_[np.ones((M, 1)), x1]
a_opt = np.linalg.pinv(np.transpose(X).dot(X)).dot(np.transpose(X).dot(y))
yhat = a_opt[0,0] + a_opt[1,0]*x1
Joptimum = (1/M)*np.sum(np.power((y-yhat),2))
print(f'Os valores de a0 e a1 encrontrados pela equação normal são: {a_opt[0,0]}; {a_opt[1,0]}')
```

```python
# implementação odo gradiente descedente
maxNumIter = 10000

alpha = 0.1

a = np.zeros((2,maxNumIter))
Jgd = np.zeros(maxNumIter)

a[0,0] = -10
a[1,0] = -10

yhat = a[0,0] + a[1,0]*x1

Jgd[0] =    (1/M)*np.sum(np.power((y-yhat),2))

error = 1
iter = 0

while(error>0.001 and iter<maxNumIter):
    yhat = a[0, iter] + a[1, iter]*x1
    update = -(2/M)*np.transpose((y-yhat)).dot(X)
    a_aux = a[:,iter] - alpha*np.transpose(update)
    a[0,iter+1] = a_aux[0,0]
    a[1,iter+1] = a_aux[1,0]
    yhat = a[0,iter+1] + a[1,iter+1]*x1
    Jgd[iter+1] = (1/M)*sum(np.power(y-yhat,2))
    error = np.abs(Jgd[iter]-Jgd[iter+1])
    iter =iter+1


```

```python
fig = plt.figure(figsize=(10,10))

cp = plt.contour(A0, A1, J)
plt.clabel(cp, inline=1, fontsize=10)
plt.xlabel('$a_0$')
plt.ylabel('$a_1$')
plt.title('Cost-function\'s Contour')

plt.plot(a_opt[0], a_opt[1], c='r', marker='*')
plt.plot(a[0, 0:iter], a[1, 0:iter], 'kx')

plt.xticks(np.arange(-10, 14, step=4.0))
plt.yticks(np.arange(-10, 14, step=4.0))
plt.show()
```
